```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

n_deg = 3
X = np.array([1, 3, 7, 10, 15, 25, 50]).reshape(-1, 1)
t = np.array([1, 10, 15, 22, 0, 20, 70]).reshape(-1, 1)

polynomial_features = PolynomialFeatures(degree=n_deg)
X_poly = polynomial_features.fit_transform(X)

polynomial_regression = LinearRegression()
polynomial_regression.fit(X_poly, t)

X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_range_poly = polynomial_features.transform(X_range)
t_pred = polynomial_regression.predict(X_range_poly)

plt.plot(X, t, color='red')
plt.plot(X_range, t_pred, color='blue', label=f'Polinômio Grau {n_deg}')
plt.show()
```
